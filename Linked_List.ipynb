## Single Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.addr = None

class Sinle_Linked_List:
    def __init__(self):
        self.head = None

    def insert_at_last(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            temp = self.head
            while temp.addr != None:
                temp = temp.addr
            temp.addr = newNode

    def insert_at_first(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            newNode.addr = self.head
            self.head = newNode

    def insert_at_loc(self,val,loc):
        newNode = Node(val)
        if loc <= 0:
            print("Enter location Above 0")
        elif loc == 1:
            newNode.addr = self.head
            self.head = newNode
        elif loc == self.length()+1:
            self.insert_at_last(val)
        elif loc > self.length():
            print("enter less than location",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.addr != None and cnt < loc-1:
                temp = temp.addr
                cnt +=1
            newNode.addr = temp.addr
            temp.addr = newNode

    def display(self):
        if self.head is None:
            print("No Node to Display")
        else:
            temp = self.head
            while temp:
                print(temp.data, end=" -> ")
                temp = temp.addr
            print()

    def length(self):
        if self.head is None:
            print("No Node to COunt")
        else:
            temp = self.head
            cnt = 0
            while temp:
                cnt += 1
                temp = temp.addr
            return cnt
    def delete_at_first(self):
        if self.head is None:
            print("No Node to Remove")
        else:
            self.head = self.head.addr

    def delete_at_last(self):
        if self.head is None:
            print("No Node to Remove")
        elif self.length() == 1:
            self.head = None
        else:
            temp = self.head
            while temp.addr != None:
                temp = temp.addr
            temp.addr = None 

    def delete_at_loc(self,loc):
        if loc <= 0:
            print("Enter location Above 0")
        elif loc == 1:
            self.head = self.head.addr
        elif loc > self.length():
            print("enter less than location",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.addr != None and cnt < loc-1:
                temp = temp.addr
                cnt +=1
            temp.addr = temp.addr.addr

S1 = Sinle_Linked_List()
while True:
    print("----Single Linked List----")
    print("1 - insert at last\n2 - insert at first\n3 - insert at loc\n4 - display\n5 - length\n" \
            "6 - delete at last\n7 - delete at first\n8 - delete at loc")
    opt = int(input("Enter Your Option:"))
    match opt:
        case 1:
            val = input("Enter the value")
            S1.insert_at_last(val)
        case 2:
            val = input("Enter the value")
            S1.insert_at_first(val)
        case 3:
            val = input("Enter the value")
            loc = int(input('Enter the loc'))
            S1.insert_at_last(val)
        case 4:
            S1.display()
        case 5:
            print("Totel Nodes - ",S1.length())
        case 6:
            S1.delete_at_last()
        case 7:
            S1.delete_at_first()
        case 8:
            loc = int(input('Enter the loc'))
            S1.delete_at_loc()
        case _:
            print("invalid Option")
```

## Circular Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.addr = None

class Circular_Linked_List:
    def __init__(self):
        self.head = None

    def insert_at_last(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
            self.head.addr = newNode
        else:
            temp = self.head
            while temp.addr != self.head:
                temp = temp.addr
            temp.addr = newNode
            newNode.addr = self.head

    def insert_at_first(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
            self.head.addr = newNode
        else:
            temp = self.head
            while temp.addr != self.head:
                temp = temp.addr
            temp.addr = newNode
            newNode.addr = self.head
            self.head = newNode

    def insert_at_loc(self,val,loc):
        newNode = Node(val)
        if loc <= 0:
            print("Enter location Above 0")
        elif loc == 1:
            self.insert_at_first(val)

        elif loc == self.length()+1:
            self.insert_at_last(val)
        elif loc > self.length():
            print("enter less than location",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.addr != self.head and cnt < loc-1:
                temp = temp.addr
                cnt +=1
            newNode.addr = temp.addr
            temp.addr = newNode

    def display(self):
        if self.head is None:
            print("No Node to Display")
        else:
            temp = self.head
            while temp:
                print(temp.data, end=" -> ")
                temp = temp.addr
                if temp == self.head:
                    break
            print()

    def length(self):
        if self.head is None:
            print("No Node to COunt")
        else:
            temp = self.head
            cnt = 0
            while temp:
                cnt += 1
                temp = temp.addr
                if temp == self.head:
                    break
            return cnt
        
    def delete_at_first(self):
        if self.head is None:
            print("No Node to Remove")
        else:
            self.head = self.head.addr

    def delete_at_last(self):
        if self.head is None:
            print("No Node to Remove")
        # elif self.length() == 1:
        #     self.head = self.head
        else:
            temp = self.head
            while temp.addr.addr != self.head:
                temp = temp.addr
            temp.addr = self.head 

    def delete_at_loc(self,loc):
        if loc <= 0:
            print("Enter location Above 0")
        elif loc == 1:
            self.head = self.head.addr
        elif loc > self.length():
            print("enter less than location",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.addr != None and cnt < loc-1:
                temp = temp.addr
                cnt +=1
            temp.addr = temp.addr.addr

C1 = Circular_Linked_List()
while True:
    print("----Circular Linked List----")
    print("1 - insert at last\n2 - insert at first\n3 - insert at loc\n4 - display\n5 - length\n" \
            "6 - delete at last\n7 - delete at first\n8 - delete at loc")
    opt = int(input("Enter Your Option:"))
    match opt:
        case 1:
            val = input("Enter the value")
            C1.insert_at_last(val)
        case 2:
            val = input("Enter the value")
            C1.insert_at_first(val)
        case 3:
            val = input("Enter the value")
            loc = int(input('Enter the loc'))
            C1.insert_at_last(val)
        case 4:
            C1.display()
        case 5:
            print("Totel Nodes - ",S1.length())
        case 6:
            C1.delete_at_last()
        case 7:
            C1.delete_at_first()
        case 8:
            loc = int(input('Enter the loc'))
            C1.delete_at_loc()
        case _:
            print("invalid Option")
```

## using stack to create a linked list

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.addr = None

class Stack_Linked_List:
    def __init__(self):
        self.head = None
    
    def display(self):
        if self.head is None:
            print("No Nodes to Display")
        else:
            temp = self.head
            while temp :
                print(temp.data)
                temp = temp.addr
            print()
        
    
    def push(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            newNode.addr = self.head
            self.head = newNode

    def pop(self):
        if self.head is None:
            print("No nodes to Remove")
        else:
            self.head = self.head.addr

    def peek_ele(self):
        if self.head is None:
            print("No element to peek")
        else:
            print("Peek element is -> ",self.head.data)

S1 = Stack_Linked_List()
S1.push(10)
S1.push(20)
S1.push(30)
S1.push(40)
S1.display()
```

## Double Linked List

```python
class Node:
    def __init__(self,data):
        self.prev = None
        self.data = data
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_last(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            temp = self.head
            while temp.next != None:
                temp = temp.next
            temp.next = newNode
            newNode.prev = temp
    
    def display(self):
        if self.head is None:
            print("No nodes to Display")
        else:
            temp = self.head
            while temp:
                print(temp.data, end=" <=> ")
                temp = temp.next
            print()

    def length(self):
        if self.head is None:
            print("No nodes to count")
        else:
            temp = self.head
            cnt = 0
            while temp:
                cnt +=1
                temp = temp.next
            return cnt

    def insert_at_first(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            newNode.next = self.head
            self.head.prev = newNode
            self.head = newNode

    def insert_at_loc(self,val,loc):
        newNode = Node(val)
        if loc <=0:
            print("Enter Loction Above 0")
        elif loc == 1:
            self.insert_at_first(val)
        elif loc == self.length()+1:
            self.insert_at_last(val)
        elif loc > self.length():
            print("Enter loc less than ",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.next != self.head and cnt < loc -1:
                temp = temp.head
                cnt +=1
            temp.next.prev = newNode
            newNode.next = temp.next
            newNode.prev = temp
            temp.next = newNode

    def delete_at_last(self):
        if self.head is None:
            print("No Node to Remove")
        elif self.length() == 1:
            self.head = None
        else:
            temp = self.head
            while temp.next.next != None:
                temp = temp.next
            temp.next.prev = None
            temp.next = None
    def delete_at_first(self):
        if self.head is None:
            print("No Nodes to Remove")
        elif self.length() == 1:
            self.head = None
        else:
            self.head.next.prev = None
            self.head = self.head.next

    def delete_at_loc(self,loc):
        if self.head is None:
            print("No Nodes to remove")
        elif loc <= 0:
            print("Enter more than loc 0")
        elif loc == self.length():
            self.delete_at_last()
        elif loc == 1:
            self.delete_at_last()
        elif loc > self.length():
            print("give less than",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.next != None and cnt < loc-1:
                temp = temp.next
                cnt += 1
            temp.next = temp.next.next
            temp.next.prev = temp

    def search_the_element(self,val):
        if self.head is None:
            print("No Node")
        else:
            temp = self.head
            while temp.next != None:
                if temp.data == val:
                    print("Value Found")
                    if temp.next is None:
                        print("The next node is None")
                    else:
                        print("The Next Node is",temp.next.data)
                    if temp.prev is None:
                        print("The No Previous Node")
                    else:
                        print("The Previous Node is",temp.prev.data)
            else:
                print("Value Not Found")

cll = DoubleLinkedList()
cll.insert_at_last(10)
cll.insert_at_last(20)
cll.insert_at_last(30)
cll.insert_at_last(40)
cll.insert_at_last(50)
cll.display()
cll.insert_at_first(60)
cll.display()

```

## Circular Double Linked List
