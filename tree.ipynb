```python
class TREE:
    def __init__(self,data):
        self.prev = None
        self.data = data
        self.next = None

    def insert_the_ele(self,val):
        if self.data:
            if val < self.data:
                if self.prev is None:
                    self.prev = TREE(val)
                else:
                    self.prev.insert_the_ele(val)
            elif val > self.data:
                if self.next is None:
                    self.next = TREE(val)
                else:
                    self.next.insert_the_ele(val)

    def display(self):
        if self.prev:
            self.prev.display()
        print(self.data, end=" -> ")
        if self.next:
            self.next.display()

    def pre_order_traversal(self,root):
        if root:
            print(root.data, end=" -> ")
            self.pre_order_traversal(root.prev)
            self.pre_order_traversal(root.next)

    def in_order_traversal(self,root):
        if root:
            self.in_order_traversal(root.prev)
            print(root.data,end=" -> ")
            self.in_order_traversal(root.next)

    def post_order_traversal(self,root):
        if root:
            self.post_order_traversal(root.prev)
            self.post_order_traversal(root.next)
            print(root.data,end=" -> ")

    def bredth_first_search(self,root):
        if root:
            queue = [root]
            while queue:
                del_ele = queue.pop(0)
                print(del_ele.data, end=" -> ")
                if del_ele.prev:
                    queue.append(del_ele.prev)
                if del_ele.next:
                    queue.append(del_ele.next)

    def search_the_ele(self,val):
        if val < self.data:
            if self.prev is None:
                return "Value Not Fount"
            else:
                return self.prev.search_the_ele(val)
        elif val > self.data:
            if self.next is None:
                return "Value Not Found"
            else:
                return self.next.search_the_ele(val)
        else:
            return "Value Not Found"

    def height_of_tree(self,root):
        if root is None:
            return 0
        leftheight = self.height_of_tree(root.prev)
        rightheight = self.height_of_tree(root.next)
        if leftheight > rightheight:
            return leftheight+1
        return rightheight+1

t = TREE(15)
t.insert_the_ele(12)
t.insert_the_ele(20)
t.insert_the_ele(9)
t.insert_the_ele(14)
t.insert_the_ele(18)
t.insert_the_ele(25)
t.insert_the_ele(6)
t.insert_the_ele(10)
print("Pre Order")
t.pre_order_traversal(t)
print()
print("In Order")
t.in_order_traversal(t)
print()
print("Post Order")
t.post_order_traversal(t)
print()
print("ZigZog Order")
t.bredth_first_search(t)
print()
print("the height of tree",t.height_of_tree(t))

```
