```python
# stack using without limit
class STACK:
    def __init__(self):
        self.s1 = []
    
    def display(self):
        if len(self.s1) != 0:
            for i in range(len(self.s1)-1,-1,-1):
                print(self.s1[i])
        else:
            print("No Elements in Stack")
    
    def push(self,val):
        self.s1.append(val)
    
    def pop(self):
        self.s1.pop()
    
    def peek_ele(self):
        print("Top element in stack:",self.s1[-1])

st = STACK()
st.push(10)
st.push(20)
st.push(30)
st.push(40)
st.pop()
st.display()
    
```

```python
# stack using limit
class STACK:
    def __init__(self,maxSize):
        self.s1 = []
        self.maxSize = maxSize
    
    def isFull(self):
        if len(self.s1) == self.maxSize:
            return True
        else:
            return False
    
    def isEmpty(self):
        if len(self.s1) == 0:
            return True
        else:
            return False 
    
    def display(self):
        if self.isEmpty():
            print("No Elements in Stack")
        else:
            for i in range(len(self.s1)-1,-1,-1):
                print(self.s1[i])
    
    def push(self,val):
        if self.isFull():
            print("Stack is Full")
        else:
            self.s1.append(val)
    
    def pop(self):
        if self.isEmpty():
            print("Stack has no elements")
        else:
            self.s1.pop()
    
    def peek_ele(self):
        if self.isEmpty():
            print("stack has no elements")
        else:
            
            print("Top element in stack:",self.s1[-1])

st = STACK(5)
while True:
    print("------STACK OPERATION------")
    print("1. Push \n2. Pop \n3. Display \n4.Peek \n5.isFull \n6. isEmpty")
    opt = int(input("Enter your option : "))
    match opt:
        case 1:
            val = int(input("Enter the Element : "))
            st.push(val)
        case 2:
            st.pop()
        case 3:
            st.display()
        case 4:
            st.peek_ele()
        case 5:
            print("Is stack is full :",st.isFull())
        case 6:
            print("is stock is empty",st.isEmpty())
        case _:
            print("Enter valid option")
            exit()
```

```python
def balanced_paranthesis(s1):
    stack = []
    values = { '{':'}','[':']','(':')'}

    for char in s1:
        if char in "[{(":
            stack.append(char)
        elif char in "]})":
            if not stack or stack.pop() != values[char]:
                return False
    return len(stack) == 0


s1='[]{}()'

if balanced_paranthesis(s1):
    print('Balanced')
else:
    print("Not Balanced")
```
