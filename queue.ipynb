**Linear Queue**

```python
class  QUEUE:
    def __init__(self,maxsize):
        self.q1 = []
        self.maxsize = maxsize
        
    def isfull(self):
        if len(self.q1) == self.maxsize :
            return True
        else:
            return False
        
    def isempty (self):
        if len(self.q1) == 0:
            return True
        else:
            return False
        
    def enqueue (self,val):
        if self.isfull():
            print("Queue is full")
            
        else:
            self.q1.append(val)
            
    def peek(self):
        if self.isempty():
            print("No element to peek")
            
        else:
            print("Peek element : ",self.q1[0])
            
    def display(self):
        if self.isempty():
            print("No elements in queue")
        else:
            print("Elements in queue : ",self.q1)
            
            
    def dequeue(self):
        if self.isempty():
            print("No element to delete")
        else:
            print("Deleted element : ",self.q1.pop(0))
            
            
q = QUEUE(5)
q.enqueue('A')
q.enqueue('B')
q.enqueue('C')
q.peek()
q.display()
q.dequeue()
q.display()
q.peek()
```

## Circular Queue

```python
class Circular_Queue:
    def __init__ (self,maxsize):
        self.c1 = [None] * maxsize
        self.maxsize = maxsize
        self.front = -1
        self.rear = -1
        
    def isfull(self):
        if self.front == 0 and self.rear == self.maxsize -1 :
            return True
        elif self.rear +1 == self.front: 
            return True
        else:
            return False
        
    def isempty (self):
        if self.front == -1 and self.rear ==-1:
            return True
        else:
            return False
        
    def display(self):
        print(self.c1)
        
    def enqueue(self,val):
        if self.isfull():
            print("Circular queue is full")
        else:
            if self.rear +1 == self.maxsize:
                self.rear = 0
            else:
                self.rear += 1
                if self.front == -1:
                    self.front = 0
            self.c1[self.rear] = val
            
    def peek(self):
        if self.isempty():
            print("No element to peek")
        else:
            print("Peek element : ",self.c1[self.front])
            
    def dequeue(self):
        idx =self.front
        if self.front == self.rear:
            self.front == -1
            self.rear == -1
            
        elif self.front + 1 == self.maxsize:
            self.front =0
            
        else:
            self.front += 1 
        self.c1[idx] = None
            

c = Circular_Queue(5)
c.display()
c.isfull()
c.isempty()
c.enqueue(10)
c.enqueue(20)
c.enqueue(30)
c.enqueue(40)
c.enqueue(50)
c.display()
c.enqueue(60)
c.dequeue()


```

## Stack And Queue

```python
class Stack:
    def __init__(self):
        self.s1 = []

    def push(self,val):
        self.s1.append(val)

    def display(self):
        for i in range(len(self.s1)-1,-1,-1):
            print(self.s1[i])

    def pop(self):
        if len(self.s1) == 0:
            return None
        return self.s1.pop()

class QueueStack:
    def __init__(self):
        self.s1 = Stack()
        self.s2 = Stack()

    def enqueue(self,val):
        self.s1.push(val)

    def dequeue(self):
        while len(self.s1):
            self.s2.push(self.s1.pop())
        res = self.s2.pop()
        while len(self.s2.s1):
            self.s1.push(self.s2.pop())
        return res

    def display(self):
        self.s1.display()

qs = QueueStack()
qs.enqueue('A')
qs.enqueue('B')
qs.enqueue('C')
qs.enqueue('D')
qs.display()
print("removed element - ",qs.dequeue())
```
