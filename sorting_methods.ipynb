## Buble Sort

```python
def BubbleSort(l):
    for i in range(len(l)):
        for j in range(len(l)-i-1):
            if l[j] > l[j+1]:
                l[j], l[j+1] = l[j+1], l[j]

l = [9,8,7,4,4,6,5,4,3,2,1]

print("Before Sort -", l)
BubbleSort(l)
print("After Sort -", l)
```

## Selection Sort

```python
def SelectionSort(l):
    for i in range(len(l)):
        min_ele = i
        for j in range(i+1,len(l)):
            if l[min_ele] > l[j]:
                min_ele = j
        l[i], l[min_ele] = l[min_ele],l[i]
l = [9,8,7,6,5,4,3,2,1]
print("Before Sort - ",l)
SelectionSort(l)
print("After Sort - ",l)
```

## Quick Sort

```python
def partition(start,end,l1):
    i = start
    j = end
    pivot = l1[end]
    while i < j:
        while l1[i] < pivot:
            i+=1
        while l1[j] > pivot:
            j-=1
        if i<j:
            l1[i], l1[j] = l1[j],l1[i]
    if l1[i] > pivot:
        l1[i],l1[end] = l1[end],l1[i]
    return i
```

```python
def Quick_Sort(start,end,l1):
    if start < end:
        pi = partition(start,end,l1)
        Quick_Sort(start,pi-1,l1)
        Quick_Sort(pi+1,end,l1)

l1 = [3,5,7,1,4,8,6]
print("Before Sort - ",l1)
Quick_Sort(0,len(l1)-1,l1)
print("After Sorting - ",l1)
```

## Merge Sort

```python
def merge_sort(l1):
    mid = len(l1)//2
    left_half = l1[:mid]
    right_half = l1[mid:]

    i = 0 # left side
    j = 0 # rigth side
    k = 0 # original list
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            l1[k] = left_half[i]
            i +=1
        else:
            l1[k] = right_half[j]
            j +=1
        k +=1
    while i < len(left_half):
        l1[k] = left_half[i]
        i +=1
        k +=1
    while j < len(right_half):
        l1[k] = right_half[j]
        j +=1
        k +=1

l1 = [2,4,9,6,5,7,8]
print("Before Sorting -> ",l1)
merge_sort(l1)
print("After Sorting -> ",l1)
```

## Merge Two sorted list

```python
def merge_sort(l1,l2):
    i = 0
    j = 0 
    res = []
    while i <len(l1) and j < len(l2):
        if l1[i] < l2[j]:
            res += [l1[i]]
            i +=1
        else:
            res +=[l2[j]]
            j+=1
    while i < len(l1):
        res += [l1[i]]
        i+=1
    while j < len(l2):
        res += [l2[j]]
        j+=1
    return res

l1 = [-1,4,7,8,11]
l2 = [0,2,3,4,6,9,14]
print("Before Sort ->",l1,l2)
print("After Sort -> ",merge_sort(l1,l2))
```
