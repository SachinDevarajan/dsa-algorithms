**Factorial number**

```python
def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

**Sum of 10  natural Number**

```python
def sum(n):
    if n == 1:
        return 1
    else:
        return n + sum  (n-1)
    
print(sum(10))
```

**Sum of given number using Recurssion**

```python
def sum_if_given(n):
    if n == 0:
        return 0
    
    else:
        return(n%10) + sum_if_given(n//10)
        
print(sum_if_given(86214)) 
```

**Sum of give list**

```python
l = [5,2,1,7,3]
def sum_list(l):
    if len(l) == 0:
        return 0
    else:
        return l[0]+ sum_list(l[1:])
        
print(sum_list(l))
```

*product of give list*

```python
l = [5,2,1,7,3]

def product (l):
    if len(l) == 0 :
        return 1
    else:
        return l[0] * product(l[1:])
    
    
print(product(l))
    
```

```python
def reverse(n):
    if len(n) == 0:
        return ""
    else:
        return n[-1] + reverse(n[:-1])
    
print(reverse("pyspiders"))
        
```

```python
def palindrome(n):
    if len(n) == 0:
        return "String is Palinrome"
    elif n[0] != n[-1]:
        return "No palindrome"
    else:
        return palindrome(n[1:-2])
    
print(palindrome("mam"))
```

convert first charecter as upper and last charecter also upper in each word in a string using recursion

```python
def convert(s, i=0):
    if i == len(s):
        return ""

    elif s[i] == " ":
        return " " + convert(s, i + 1)

    elif i == 0 or s[i - 1] == " ":
        return s[i].upper() + convert(s, i + 1)

    elif i == len(s) - 1 or s[i + 1] == " ":
        return s[i].upper() + convert(s, i + 1)
    

    return s[i] + convert(s, i + 1)


s = "hi this is recursion function"
print(convert(s))
```
